### **Telco Customer Churn Prediction - Data Preparation**

**Objective**  
To develop a machine learning classification model that predicts customers who are likely to discontinue their telecom subscription services. The project aims to identify the key factors influencing customer churn through data cleaning, feature engineering, model training, and evaluation, enabling telecom companies to implement targeted customer retention strategies and reduce revenue loss.

**Notebook Objective:** Clean and preprocess customer subscription data

**Import Libraries**

````python
import pandas as pd
import numpy as np
import os
````

**Load and Inspect Data**

````python
data = pd.read_csv("C:\\Users\\KOLADE\\OneDrive\\Documents\\AkoladeDSJourney\\Telco-Customer-Churn-Prediction\\data\\raw\\WA_Fn-UseC_-Telco-Customer-Churn.csv")
df = data.copy()
df.info()
````

````python
df.head()
````

````python
print(f"Number of duplicate rows: {df.duplicated().sum()}") # check for duplicates
print(f"Number of missing values: {df.isnull().sum().sum()}") # check for missing values
````

- No duplicates record
- No explicit missing values
- Incorrect Data Types: `TotalCharges` stored as Object

````python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors='coerce') # convert TotalCharges to numeric

print(f"TotalCharges Dtype: {df['TotalCharges'].dtype}") # check TotalCharges dtype
print(f"Number of missing values in TotalCharges: {df['TotalCharges'].isnull().sum()}") # check for missing values in TotalCharges
````

````python
df[df["TotalCharges"].isnull()] # display rows with missing TotalCharges
````

````python
df["TotalCharges"].describe() # check for outliers in TotalCharges
````

````python
df["TotalCharges"].isna().mean() # check percentage of missing values in TotalCharges
````

Though missing row in `TotalCharges` is small - it's safe to drop them, instead of dropping it will be filled with the median.

````python
df["TotalCharges"].fillna(df["TotalCharges"].median(), inplace=True) # fill missing TotalCharges with median

print(f"Number of missing values in TotalCharges after filling: {df['TotalCharges'].isnull().sum()}") # check for missing values in TotalCharges after filling
````

````python
print(f"Number of duplicate rows: {df.duplicated().sum()}") # check for duplicates
print(f"Number of missing values: {df.isnull().sum().sum()}") # check for missing values
````

````python
print("Data preparation completed successfully.")
````
