**Load datasets**

```python
original_results = pd.read_csv("C:\\Users\\KOLADE\\OneDrive\\Documents\\AkoladeDSJourney\\Telco-Customer-Churn-Prediction\\reports\\results\\original_results.csv")
original = original_results.copy()

cw_results = pd.read_csv("C:\\Users\\KOLADE\\OneDrive\\Documents\\AkoladeDSJourney\\Telco-Customer-Churn-Prediction\\reports\\results\\cw_results.csv")
cw = cw_results.copy()

smote_results = pd.read_csv("C:\\Users\\KOLADE\\OneDrive\\Documents\\AkoladeDSJourney\\Telco-Customer-Churn-Prediction\\reports\\results\\smote_results.csv")
smote = smote_results.copy()
```

```python
results = pd.concat([original, cw, smote], axis=0, ignore_index=True)

results.sort_values(by=['Recall', 'F1', 'ROC_AUC'], ascending=False, inplace=True)

display(results)
```

#### **Model Performance Summary**

Three imbalance-handling strategies were evaluated:

1. Original training data without intervention.
2. Original data with algorithm-level balancing techniques (`class_weight='balanced'` and `scale_pos_weight`).
3. SMOTE-based oversampling on the training data.

The results showed that imbalance-handling effectiveness varied across algorithms.

Logistic Regression achieved the strongest overall performance when combined with class weighting, producing the highest Recall (82.9%) and F1-score (64.7%) while maintaining a high ROC-AUC (85.5%). This makes it particularly suitable for churn prediction, where identifying customers likely to leave is more important than minimizing false positives (*Recall > Precision* - the usual trade-off).

XGBoost benefited substantially from `scale_pos_weight`, outperforming its SMOTE counterpart and demonstrating that algorithm-level methods can be more effective than synthetic oversampling for boosting algorithms.

Random Forest and Decision Tree models showed modest improvements under SMOTE, though their overall performance remained below that of Logistic Regression and XGBoost.

Overall, the findings suggest that algorithm-level imbalance handling is sufficient for moderately imbalanced datasets such as Telco Churn, and that synthetic oversampling should be validated empirically rather than assumed to improve performance.
