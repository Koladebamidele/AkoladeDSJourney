### **Telco Customer Churn Prediction - Class Imbalance**

**Objective:** To develop a machine learning classification model that predicts customers who are likely to discontinue their telecom subscription services. The project aims to identify the key factors influencing customer churn through data cleaning, feature engineering, model training, and evaluation, enabling telecom companies to implement targeted customer retention strategies and reduce revenue loss.

**Guiding Principles for this Notebook:** 

+ Import Libraries
+ Load Data
+ Check Target Distributions
+ Define the Functions
+ Apply the functions

**Import Libraries**

```python
import pandas as pd
import numpy as np
from imblearn.over_sampling import SMOTE
```

**Load Data**

```python
data = pd.read_parquet("C:\\Users\\KOLADE\\OneDrive\\Documents\\AkoladeDSJourney\\Telco-Customer-Churn-Prediction\\data\\processed\\train_processed.parquet")
train_df = data.copy()

train_df.head()
```

```python
# separate features and target variable
X_train = train_df.drop('Churn', axis=1)
y_train = train_df['Churn']

print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
# check the class distribution
print("\nClass distribution before SMOTE:")
print(y_train.value_counts())
```

**Define the Functions** 

```python
# Check the class distribution before applying SMOTE
def check_class_distribution(y):
    class_counts = y.value_counts()
    percentages = (class_counts / len(y)) * 100
    summary = pd.DataFrame({
        'Class': class_counts.index.map({0: 'No Churn', 1: 'Churn'}),
        'Count': class_counts.values,
        'Percentage': [f"{np.round(p, 2)}%" for p in percentages.values]
    })
    return summary
```

```python
# apply SMOTE to balance the classes
def apply_smote(X, y, random_state=42):
    smote = SMOTE(random_state=random_state)
    x_smote, y_smote = smote.fit_resample(X, y)
    return x_smote, y_smote
```

```python
# save the balanced dataset to parquet files
def save_smote_data(X, y, filepath):
    balanced_df = pd.DataFrame(X, columns=X.columns)
    balanced_df['Churn'] = y
    balanced_df.to_parquet(filepath, index=False)
    print(f"Balanced dataset saved to {filepath}")
    return balanced_df
```

**Apply the Functions**

```python
# check the class distribution before applying SMOTE
print("\nClass distribution before SMOTE:")
check_class_distribution(y_train)
```

```python
# apply SMOTE to balance the classes
X_train_smote, y_train_smote = apply_smote(X_train, y_train)

# check the class distribution after applying SMOTE
print("\nClass distribution after SMOTE:")
check_class_distribution(y_train_smote)
```

```python
# save the balanced dataset to parquet files
filepath = "C:\\Users\\KOLADE\\OneDrive\\Documents\\AkoladeDSJourney\\Telco-Customer-Churn-Prediction\\data\\processed\\train_smote.parquet"
save_smote_data(X_train_smote, y_train_smote, filepath)
```
