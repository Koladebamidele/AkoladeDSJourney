### **Telco Customer Churn Prediction - Modeling**

**Objective:** To develop a machine learning classification model that predicts customers who are likely to discontinue their telecom subscription services. The project aims to identify the key factors influencing customer churn through data cleaning, feature engineering, model training, and evaluation, enabling telecom companies to implement targeted customer retention strategies and reduce revenue loss.

**Guiding Principles for this Notebook:** 

+ Import Libraries
+ Load Data
+ Train Original train dataset
+ Train train dataset with class weights
+ Train the SMOTE train dataset
+ Save results as csv

**Import Libraries**

```python
import pandas as pd

from src.modeling import get_models
from src.evaluation import evaluate_models
```

**Load Data**

```python
# import the processed train and test datasets
train_data = pd.read_parquet("C:\\Users\\KOLADE\\OneDrive\\Documents\\AkoladeDSJourney\\Telco-Customer-Churn-Prediction\\data\\processed\\train_processed.parquet")
train_df = train_data.copy()
test_data = pd.read_parquet("C:\\Users\\KOLADE\\OneDrive\\Documents\\AkoladeDSJourney\\Telco-Customer-Churn-Prediction\\data\\processed\\test_processed.parquet")
test_df = test_data.copy()

# import the balanced train dataset after applying SMOTE
smote_train_data = pd.read_parquet("C:\\Users\\KOLADE\\OneDrive\\Documents\\AkoladeDSJourney\\Telco-Customer-Churn-Prediction\\data\\processed\\train_smote.parquet")
smote_train_df = smote_train_data.copy()

print(f"Train dataset shape: {train_df.shape}")
print(f"SMOTE balanced train dataset shape: {smote_train_df.shape}")
print(f"Test dataset shape: {test_df.shape}")
```

**Split Data**

```python
# Split the original train dataset into features and target variable
X_train = train_df.drop(columns=['Churn'])
y_train = train_df['Churn']

# Split the test dataset into features and target variable
X_test = test_df.drop(columns=['Churn'])
y_test = test_df['Churn']

# Split the SMOTE balanced train dataset into features and target variable
X_train_smote = smote_train_df.drop(columns=['Churn'])
y_train_smote = smote_train_df['Churn']
```

**Build models with original train dataset**

```python
models = get_models(y_train = y_train)
print(f"Models to be evaluated: {list(models.keys())}")
```

```python
# evaluate the models on the original train dataset
original_results = evaluate_models(
    models, 
    X_train, 
    y_train, 
    X_test, 
    y_test, 
    experiment_name="original_train"
)

original_results
```

**Build models with original train datasets + class weight**

```python
cw_models = get_models(
    y_train = y_train, 
    use_class_weights = True
)

print(f"Models with class weights to be evaluated:\n {list(cw_models.keys())}")
```

```python
# evaluate the models on the original + class weights train dataset
cw_results = evaluate_models(
    cw_models, 
    X_train, 
    y_train, 
    X_test, 
    y_test, 
    experiment_name="original + class weights"
)

cw_results
```

**Build models with the SMOTE train datasets**

```python
smote_models = get_models(
    y_train = y_train_smote
)
print(f"Models with SMOTE to be evaluated:\n {list(smote_models.keys())}")
```

```python
# evaluate the models on the SMOTE balanced train dataset
smote_results = evaluate_models(
    smote_models, 
    X_train_smote, 
    y_train_smote, 
    X_test, 
    y_test, 
    experiment_name="SMOTE"
)

smote_results
```

**Save results**

```python
# save the results to CSV files
original_results.to_csv("C:\\Users\\KOLADE\\OneDrive\\Documents\\AkoladeDSJourney\\Telco-Customer-Churn-Prediction\\reports\\results\\original_results.csv", index=False)
print(r"Original results saved to ..reports\\results as CSV.")

cw_results.to_csv("C:\\Users\\KOLADE\\OneDrive\\Documents\\AkoladeDSJourney\\Telco-Customer-Churn-Prediction\\reports\\results\\cw_results.csv", index=False)
print(r"Class weights results saved to ..reports\\results as CSV.")

smote_results.to_csv("C:\\Users\\KOLADE\\OneDrive\\Documents\\AkoladeDSJourney\\Telco-Customer-Churn-Prediction\\reports\\results\\smote_results.csv", index=False)
print(r"SMOTE results saved to ..reports\\results as CSV.")
```
